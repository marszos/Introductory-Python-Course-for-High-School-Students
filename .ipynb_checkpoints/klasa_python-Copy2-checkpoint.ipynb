```python
# co to jest slownik? 
# Słownik kojarzy klucze (nazwy) z wartościami (szczegóły). Stąd używana czasem nazwa ‘tablice asocjacyjne’ (od asocjacja = skojarzenie). 
# Należy pamiętać, że klucz musi być unikalny. np. takim kluczem moglby byc numer pesel 
# Klucze w słowniku są niezmienne. Zmieniać można natomiast wartości. 
# Słowniki zapisujemy w postaci par klucz:wartość, rozdzielonych przecinkami i ujętymi w nawiasy klamrowe. 
# Na przykład: d = {klucz1 : wartosc1, klucz2 : wartosc2 }‘.
# w obiekcie słownika mamy trzy metody zwane widokami, które zwracają odpowiednio listę par, 
# listę kluczy i listę wartości: dict.keys(), dict.values(),dict.items(). 
# Słownik {1:jeden,2:dwa} jest klasy dict, ale {1:je- den,2:dwa}.items() to już lista.
```

```python
# key-value pair allows users to quickly grab objects whitout needing to know an index location 
# you just call the key and we get value associated with that key 
# dictionaries use curly braces and colons to signify the keys and their associated values 
# {'key1':'value1', 'key2':'value2'}
# pytanie kiedy wybrac liste a kiedy wybrac slownik? 
# w slownikach elementy (obiekty) pobierane sa za pomoca nazwy klucza / jezeli znamy klucz a nie znamy indeksu
# slowniki sa nieuporzadkowane i nie moga byc sortowane 
# w listy pobieranie elementow za pomoca indexow / mozna indexowac i slice'owac / tracimy ta funkcjonalnosc w slowniku
```

```python
# Pęlta while

# Pętla while jest podobna do instrukcji warunkowej (if ). 
# W obu wypadkach blok instrukcji wykonuje się pod warunkiem, że podany warunek (wyrażenie logiczne) jest spełniony. 
# Jedyna różnica polega na tym, że blok warunkowy w instrukcji if wykona się raz, 
# gdy tymczasem blok pętli while wykonuje się tak długo, jak długo warunek jest spełniony.
```

```python
# spojrzmy na kilka przykladow petli while 
```

```python
# inkrementacja https://pl.wikipedia.org/wiki/Inkrementacja
# += 

# czesto wykorzystywany operator inkrementacji 
# += adds another value with the variable's value and assigns the new value to the variable.

x = 0

while x < 10:
    print('x is currently: ',x)
    print(' x is still less than 10, adding 1 to x')
    x+=1
```

```python
# powyzsza petla while zatrzymala sie kiedy wystapil warunek x == 10 
# dadajmy do tej petli wyrazenie else 
```

```python
x = 0

while x < 10:
    print('x is currently: ',x)
    print(' x is still less than 10, adding 1 to x')
    x+=1
    
else:
    print('All Done!')
```

```python
# w naszych loopach mozemy dla dodania funkcjonalnosci dla roznych przypadkow poslugiwac sie wyrazeniami 
# break, continue, pass 
```

```python
# instrukcja break sluzy do przerwania petli bez wzgledu na to czy zosataly spelnione warunki jej zakonczenia. 
# Instrukcja continue powoduje pominięcie reszty bloku instrukcji w pętli. 
# Po wykonaniu instrukcji Python przystępuje do wykonania następnej iteracji (powtórzenia) pętli. 
# pass: Does nothing at all.
```

```python
x = 0

while x < 10:
    print('x is currently: ',x)
    print(' x is still less than 10, adding 1 to x')
    x+=1
    if x==3:
        print('x==3')
    else:
        print('continuing...')
        continue
```

```python
# prosze zauwazyc, iz na wyjsciu pojawia sie x == 3 a nastepnie instrukcja jest kontynuowana. zobaczymy co sie stanie
# jak wprowadzimy break po warunku x == 3 
```

```python
x = 0

while x < 10:
    print('x is currently: ',x)
    print(' x is still less than 10, adding 1 to x')
    x+=1
    if x==3:
        print('Breaking because x==3')
        break
    else:
        print('continuing...')
        continue
```

```python
# teraz po spelnieniu warunku x==3 wyrazenie else wasn't reached i kontynuacja nie zostala wydrukowna. 
```

```python
# wazna uwaga!!! przy uzywaniu petli while moze sie zdarzyc ,iz stworzymy petle NIESKONCZONA!!!
# w celu zatrzymania petli wchodzimy w  Jupiter Notebook w opjce Kernel i albo Interrupt albo Restart!!!

while True:
    print("Utknąłem w nieskończonej pętli!")
```
